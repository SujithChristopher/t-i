Question 2: B

```python
# H(S) = S/(S + 1/1000 + S^2 * 10^-6)
# R out = 1000 Ohm
# C1 = 10^-6 F
# L1 = 1 mH

freq = np.arange(0, 1000, 0.1)
s = None
_h = []
for i in freq:
    s = 1j*i
    h_s = s/(s + 1000 + s*s*10**-6)
    _h.append(h_s)
_h = np.array(_h)
```

```python
ab = abs(_h)
# log value
lg = np.log(ab)*20

plt.figure()
plt.plot(freq, lg,"b", label="Mag vs freq")
plt.legend()
plt.show()
```

```python
plt.plot(freq, np.angle(_h), "b", label="Phase response")
plt.legend()
plt.show()
```
