Question 2:B

H(S) = S/(S + 1000 + S^2 * 10^-6), 
R out = 1000 Ohm, 
C1 = 10^-6 F, 
L1 = 1 mH, 

```python
num = [1]
den = [10**-6, 1, 10**3]
tf = signal.TransferFunction(num, den) # create transfer function
zpg = tf.to_zpk() # convert to zero-pole-gain form
w, H = signal.freqresp(zpg) # calculate frequency response
```

```python
# create subplot 1
fig, ax = plt.subplots(1,2)
ax[0].plot(w, 20*np.log(abs(H)), label="Mag vs freq")
ax[0].legend()
ax[0].set_xlabel('Frequency')
ax[0].set_ylabel('Magnitude (dB)')
# size of subplot 2
# ax[0].set_figwidth(10)
ax[1].plot(w, np.angle(H), label="Phase response")
ax[1].legend()
ax[1].set_xlabel('Frequency')
ax[1].set_ylabel('Phase')
fig.suptitle('Magnitude and Phase Response of H(s) = 1/(s^2 + 10^3s + 10^-6)')
plt.show()
```

Question 2:D, L2 = 2mH, Rout = 1k, C1 = 1 micro F,
<br>
H(S) = S/ [S^2 * (L2/Rout) + S + 1/(C1*Rout)]

```python
num = [1]
den = [(2*10**-3)/10**3, 1, 1/(10**-6*10**3)]
tf = signal.TransferFunction(num, den) # create transfer function
zpg = tf.to_zpk() # convert to zero-pole-gain form
w, H = signal.freqresp(zpg) # calculate frequency response
tf
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(w, 20*np.log(abs(H)), label="Mag vs freq")
ax[0].legend()
ax[0].set_xlabel('Frequency')
ax[0].set_ylabel('Magnitude (dB)')
# size of subplot 2
# ax[0].set_figwidth(10)
ax[1].plot(w, np.angle(H), label="Phase response")
ax[1].legend()
ax[1].set_xlabel('Frequency')
ax[1].set_ylabel('Phase')
fig.suptitle('Magnitude and Phase Response of H(s) = 1/(s^2 + 10^3s + 10^-6)')
plt.show()
```
